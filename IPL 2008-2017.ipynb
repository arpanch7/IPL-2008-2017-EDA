# Find Orange Cap Holder yearwise

```python
new=deliveries.merge(matches,left_on='match_id',right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

```python
orange=new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

```python
plt.title('Runs Scored by Orange Cap Holders')

ax=sns.barplot(x='season',y='batsman_runs',hue='batsman',width=0.8,data=orange)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

# Function to generate Orange Cap Holder yearwise

```python
def OrangeCap(Year):
    return new[new['season']==Year].groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

# Function to find out how many ducks does a batsman have


```python
def NumberOfDucks(Name):
    x=deliveries[deliveries['batsman']==Name]
    y=(x.groupby('match_id')['batsman_runs'].sum()==0).reset_index()    
    return y[y['batsman_runs']==True].shape[0]    
```

```python
NumberOfDucks('MS Dhoni')
```
